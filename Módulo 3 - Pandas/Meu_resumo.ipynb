## filtros compostos
Obs.: temos os seguintes operadores lógicos:

- &     - corresponde ao "and"
- |     - corresponde ao "or"
- ~     - corresponde ao "not"

```python
# filtrar tabela2 para media menor que 7 e frequencia >= 20
tabela2[(tabela2['Mean']<7) & (tabela2['Frequencia']>=20)]
```

## material de aprofundamento

`.cut()`  
ordena os dados, separa em bins e computa qual grupo cada linha do df pertence. o `.cut()` escolherá os bins para serem espaçados uniformemente de acordo com os proprios valores e não coma  frequência desses valores.

Parâmetros  
`pandas.cut(x, bins, right=True, labels= None, retbins=False, precision=3, include_lowest=False, duplicates='raise', ordered=True)`  
Ao informar a quantidade de grupos o pd.cut() escolhe os bins com o mesmo tamanho de janela:

## `.qcut()`

É utilizado quando queremos discretizar nossos dados em quantis. Ao informar quantidade de grupos o `pd.qcut()` escolhe os bins tal que tenhamos a mesma quantidade de valores em cada grupo.

`cut()` x `qcut()`  
- o comando `cut()` cria caixas equidistantes, mas a frequência das amostras é desigual em cada caixa  
- o comando `.qcut()` cria caixas de tamanhos desiguais, mas a frequência das amostras é igual em cada caixa

Parâmetros  
`pandas.qcut(x, q, labels=None, retbins=False, precision=3, duplicates='raise')`

## `Groupby`

Assim como no SQL, no pandas também temos um método com o qual podemos agregar os dados. `o groupby` primeiro separa os dados em grupos definidos dentro do método, aplica um tipo de operação usando agregação, transformação, filtragem ou até uma função própria e, por fim junta os resultados encontrados.

<img src="groupby.png"  style="width: 700px" >

Exemplo de aplicação da função de agregação `mean`

Utilizar o Groupbyé o mesmo que fazer a sequência:
  
1. Dividir os dados em grupos utilizando um critério
2. Aplicar uma função em cada um dos grupos separadamente
3. combinar o resultado em uma estrutura de dados

## Funções de agregação
  
com essas funções podemos aplicar operações estatísticas nos nossos dados. Exemplos:
`mean`, `std`, `max`, `min`, `count`, `sum`, `var`.  
Quando queremos aplicar apenas uma dessas operações podemos chamá-lasdiretamente após o `groupby`:

```python
tt.head()
tt.groupby(['Pclass']).Fare.mean()
#group by 'Pclass' and make mean of Fare
```

```python
tt.groupby(['Pclass'])['Fare'].mean()
```

```python
tt.groupby(['Pclass'])[['Fare']].mean()
# com dois coolchetes retorna um df
```

Opções de formatação

```python
pd.options.display.float_format='{:,.2f}'.format
```

```python
# Agrupa por Pclass e Sex e calcula a média de cada grupo
tt.groupby(['Pclass', 'Sex']).mean()
```

Para calcular a média de uma só coluna

```python
tt.groupby(['Pclass', 'Sex'])[['Age']].mean()
# tt.groupby(['Pclass', 'Sex']).mean()[['Age']]
```

```python
tt.groupby(['Pclass', 'Sex'])[['Age', 'Fare']].mean().index
```

Quando queremos mais de uma operação chamamos o método `.agg()`

```python
tt.groupby(['Pclass']).agg(['mean', 'max', 'min']).columns
```

```python
tt.groupby(['Pclass', 'Sex']).agg(['mean', 'max', 'min'])
```

Para operações distintas em colunas distintas passamos um dicionario com o nome da coluna como chave e a operação como valor

```python
tt.groupby(['Pclass']).agg(
    {'Embarked': pd.Series.mode, 'Fare':'mean'})
```

A coluna utilizada no `groupby` virou um index do nosso df. Para convertê-la em coluna novamente temos duas formas:  
  
1. chamar o parâmetro `as_index=False` dentro do `groupby`
2. aplicar `.reset_index()` ao final da sentença

```python
# método 1
tt.groupby(['Pclass'], as_index=False).agg({'Embarked': pd.Series.mode, 'Fare': 'mean'})
```

```python
# Método 2
tt.groupby(['Pclass']).agg({'Embarked':pd.Series.mode, 'Fare':'mean'}).reset_index()
```

**Ele será muito útil na criação de novas features para os modelos.**
